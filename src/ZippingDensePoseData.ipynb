```python
import os
from os.path import isfile, isdir, join, exists,getsize,basename

# The folder containing files.
directory = RESULT_FOLDER

densePoseModelName = "DensePoseKeyPointsMask_ResNet50_FPN_s1x-e2e"

# Get all files.
fileList = os.listdir(directory)

# Loop and add files to list.
pairs = []
for file in fileList:
    file = join(directory,file,densePoseModelName) 
    if not exists(file):
        continue
    # Use join to get full file path.
    location = join(directory, file)

    # Get size and add to list of tuples.
#     size = getsize(location)
    size = sum([os.path.getsize(f) for f in os.scandir(location) if os.path.isfile(f)])
    pairs.append((size, file))

# Sort list of tuples by the first element, size.
pairs.sort(reverse=True, key=lambda s: s[0])

# Display pairs.
for pair in pairs:
    print(pair)
    
# 900M	./S009C003P007R002A023_rgb
# 876G	.
# 713M	./S009C003P007R002A002_rgb 
# 664M	./S010C001P008R002A002_rgb
# 656M	./S011C002P007R001A049_rgb
# 631M	./S010C003P007R002A001_rgb
# 620M	./S003C003P017R002A001_rgb
# 616M	./S009C002P007R002A002_rgb
# 615M	./S009C001P007R002A002_rgb
# 599M	./S010C001P007R002A023_rgb
# 578M	./S002C003P003R002A023_rgb
# 571M	./S010C003P025R002A049_rgb
# 568M	./S010C003P007R001A001_rgb
# 557M	./S009C002P007R001A023_rgb
# 531M	./S010C001P007R002A001_rgb
```
